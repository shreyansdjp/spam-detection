```python
import pandas as pd
import re
```

```python
df = pd.read_csv('./data/sms-spam-collection-a-more-diverse-dataset.csv', encoding='latin-1')
df.describe()
```

```python
df.rename(columns={'sms': 'text'}, inplace=True)
df.head()
```

```python
# df.replace({'label': {'ham': 0, 'spam': 1}}, inplace=True)
df['text'] = df['text'].str.lower()
df.head()
```

```python
def remove_url(text):
    new_text = re.sub(r'(?:https?://|http//|www\.|wap\.|\.wml\b|\.com\b|\.co\b|\.tv\b|\.ac\b)', '', text)
    new_text = re.sub(r'(lt|gt|amp|quot|#39|#34|nbsp|&lt|&gt|&amp|&quot|&#39|&#34|&nbsp)', '', new_text)
    return new_text

df['text'] = df['text'].apply(remove_url)
df.head()
```

```python
df['text'] = df['text'].str.replace(r'\d{4,10}', '', regex=True)
df['text'] = df['text'].str.replace(r'to\s\d{4,}', '', regex=True)
df.head()
```

```python
df['text'] = df['text'].str.replace(r'[^A-Za-z0-9]', ' ', regex=True)
df['text'] = df['text'].str.replace(r'\s+', ' ', regex=True)
df['text'] = df['text'].str.strip()
df = df[df['text'] != '']
df.head()
```

```python
df.to_csv('./data/spam_test_preprocessed.csv', index=False)
```
